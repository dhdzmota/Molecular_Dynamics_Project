```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib notebook
```

```python
# Read data from the files

#with open('../FC/Results/initial_trial8.txt','r') as fout:
with open('../FC/initial_trial_11.txt','r') as fout:
    data_values = fout.read()

# Clean and separate data
values = data_values.split('Mbytes\n')[1].split('\nLoop time')[0].split('\n')
# Order data in a list
clean_values = [list(filter(lambda x: x !='', values[i].split(' '))) for i in range(len(values))]

# Create a Dataframe
dataframe = pd.DataFrame(clean_values[1:],columns=clean_values[0], dtype=float)
dataframe.shape
```

```python
plt.figure()
plt.plot(dataframe.Step, dataframe.Temp)
plt.grid()
#plt.ylim(200,1200)
#plt.xlim(4000,5000)
plt.xlabel('Step')
plt.ylabel('Temperature')
plt.title('Temperature')
plt.tight_layout()
```

```python
plt.figure()
plt.plot(dataframe.Step, dataframe.PotEng, label='Potential')
plt.plot(dataframe.Step, dataframe.KinEng, label='Kinetic')
plt.plot(dataframe.Step, dataframe.TotEng, label='Total')

plt.legend()
plt.grid()
plt.xlabel('Step')
plt.ylabel('Energy')
plt.title('Energies')
```

```python
plt.figure()
plt.plot(dataframe.Step, dataframe.Density)
#plt.plot(dataframe.Step, np.gradient(dataframe.Density,2))
plt.grid()
plt.xlabel('Step')
plt.ylabel('Density')
plt.title('Density')
```
